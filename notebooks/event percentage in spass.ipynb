```python
audios_path = Path('../datasets/Poliphonic mono/audios')
metadata = Path('../datasets/Poliphonic mono/meta/4_MatrixLabelsStreet.csv')
df = pd.read_csv(metadata)
df
```

```python
audio_durations = []
for audio in audios_path / df['audio_filename']:
    audio_durations.append(librosa.get_duration(filename=audio))
df['audio_duration'] = audio_durations
```

```python
df
```

```python
df['event_duration %'] = 100 * (df['end (s)'] - df['start (s)']) / df['audio_duration']
df
```

```python
df.groupby(['audio_filename', 'class']).sum()
```

```python
df.groupby(['audio_filename', 'class'])[["audio_filename", "class", "event_duration %"]].sum()
```
